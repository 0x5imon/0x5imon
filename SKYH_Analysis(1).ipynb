```python
#5 Minute Analysis of SKYH (03-17-2022 - 04-14-2022)

import pandas as pd
import requests 
import matplotlib.pyplot as plt
import numpy as np
import ta
import nbconvert
```

```python
df = pd.read_csv("SKYH_3_17_2022.csv") 

df.loc[:,'volume 1'] = df['volume'].shift(1)
df.loc[:,'V %change'] = ((df['volume 1']-df['volume'])/df['volume'])*100
df.loc[:,'MA 50'] = df['close'].rolling(window=50).mean()
df.loc[:,'MA 200'] = df['close'].rolling(window=200).mean()
df.loc[:,'EMA 9'] = df['close'].ewm(span=9, adjust=False).mean()
df.loc[:,'MACD'] = ta.trend.MACD(df['close']).macd_diff()
df.loc[:,'RSI'] = ta.momentum.RSIIndicator(df['close']).rsi()
df_slice = df.iloc[800:]
df_slice
```

```python
plt.figure(figsize=(12.2, 4.5))
plt.plot(df_slice['close'], label="Price", color='black')
#plt.plot(df_slice['EMA 9'], label="9 day EMA", color='green')
plt.plot(df_slice['MA 50'], label="50 day", color='Blue')
plt.plot(df_slice['MA 200'], label="200 Day", color='red')
plt.legend(loc = "upper left")
plt.title("Price Chart")
```

```python
plt.figure(figsize=(12.2, 4.5))
plt.plot(df_slice['MACD'])
plt.title("MACD")
```

```python
plt.figure(figsize=(12.2, 4.5))
plt.plot(df_slice['RSI'])
plt.title("RSI")
```
